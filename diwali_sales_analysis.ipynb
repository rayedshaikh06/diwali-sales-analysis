## Exploratory Data Analysis

#### Q1: Does Gender impact total purchase amount?

```python
# Total Sales amount by gender
df.groupby('Gender')['Amount'].sum().sort_values(ascending=False)
```

```python
# Viz
gender_sales = df.groupby('Gender')['Amount'].sum().sort_values(ascending=False)

ax = gender_sales.plot(kind='bar')

# Add value labels on top
for i, value in enumerate(gender_sales):
    ax.text(i, value, f'{value:,.0f}', ha='center', va='bottom')

plt.xticks(rotation=0)   # keeps M and F vertical
plt.ylabel("Total Amount")
plt.title("Total Purchase Amount by Gender")
plt.show()
```

```python
# Compute totals + percentages

gender_sales = df.groupby('Gender')['Amount'].sum()

total = gender_sales.sum()
percentages = (gender_sales / total) * 100
```

```python
# Viz
fig, ax = plt.subplots()

bars = ax.bar(gender_sales.index, gender_sales.values)

for i, bar in enumerate(bars):
    height = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f'{height:,.0f}\n({percentages.iloc[i]:.1f}%)',
        ha='center',
        va='bottom'
    )

ax.set_xlabel("Gender")
ax.set_ylabel("Total Amount")
ax.set_title("Total Purchase Amount by Gender")

plt.show()
```

#### *Insight: Female customers generate ~58% of total revenue, indicating stronger purchasing behavior.*


#### Q2: Which Age Group contributes the highest revenue?

```python
# Create age bins
bins = [0, 18, 25, 35, 45, 55, 65, 100]
labels = ['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '65+']

df['Age_Bin'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
# Display age bins, ascending by age
df['Age_Bin'].value_counts().sort_index()
```

#### *Insight: The majority of customers fall within the 26–35 age group, indicating that the platform primarily attracts young adults.*


```python
# Display age bins, descending by revenue (amount)

age_revenue = df.groupby('Age_Bin', observed=False)['Amount'].sum()
age_revenue.sort_values(ascending=False)
```

```python
# Revenue by age group

age_revenue.plot(kind='bar')
plt.xlabel("Age Group")
plt.ylabel("Total Revenue")
plt.title("Revenue by Age Group")
plt.show()
```

```python
# Viz
plt.figure()
df['Age'].plot(kind='hist', bins=20, density=True)
df['Age'].plot(kind='kde')

plt.xlabel("Age")
plt.title("Age Distribution")
plt.show()
```

#### *Insight: Customers aged 26–35 generate the highest revenue, contributing INR 42,852,589, making them the most commercially valuable age segment.*

#### Q6: Which Product Category generates the highest revenue?


```python
# Revenue by Product Category
category_revenue = df.groupby('Product_Category')['Amount'].sum()
category_revenue.sort_values(ascending=False)
```

```python
# Viz

# Get revenue sorted
category_revenue = df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False)

# Take top 10 only
top_categories = category_revenue.head(10)

plt.figure(figsize=(10,6))

bars = plt.barh(top_categories.index, top_categories.values)

plt.xlabel("Total Revenue")
plt.title("Top 10 Product Categories by Revenue")

plt.gca().invert_yaxis()  # Highest at top

# Add revenue labels
for i, value in enumerate(top_categories.values):
    plt.text(value, i, f'  {value:,.0f}', va='center')

plt.show()
```

#### *Insight: Food generates the highest revenue at INR 3.39 Cr, significantly outperforming all other categories, with Clothing & Apparel, Electronics & Gadgets, and Footwear & Shoes forming the next major revenue contributors, indicating strong concentration in essential and lifestyle segments.*


#### Q7: Which Gender prefers which Product Category?

```python
# Revenue by Gender
gender_category = df.groupby(['Gender', 'Product_Category'])['Amount'].sum().unstack()
gender_category
```

```python
# Viz

# Revenue by gender & category
gender_category = df.groupby(['Gender', 'Product_Category'])['Amount'].sum()

male_top = gender_category['M'].sort_values(ascending=False).head(8)
female_top = gender_category['F'].sort_values(ascending=False).head(8)

# --- Male Chart ---
plt.figure(figsize=(8,5))
male_top.sort_values().plot(kind='barh')
plt.title("Top Product Categories - Male")
plt.xlabel("Total Revenue")
plt.show()

# --- Female Chart ---
plt.figure(figsize=(8,5))
female_top.sort_values().plot(kind='barh')
plt.title("Top Product Categories - Female")
plt.xlabel("Total Revenue")
plt.show()
```

#### *Insight: Both male and female customers generate the highest revenue from Food, indicating it is the dominant category across genders. While spending patterns are largely similar, females show relatively stronger contribution in Footwear & Shoes, whereas males contribute slightly more toward Electronics & Gadgets, suggesting subtle preference differences within otherwise similar purchasing behavior.*








#### Q8: Does marital status influence spending?


```python
# Total revenue by Marital Status

marital_revenue = df.groupby('Marital_Status')['Amount'].sum()
marital_revenue
```

```python
# Total orders by Marital Status

marital_orders = df.groupby('Marital_Status')['Orders'].sum()
marital_orders
```

```python
# Average revenue by Marital Status

marital_avg = marital_revenue / marital_orders
marital_avg
```

```python
# Viz

marital_revenue.index = marital_revenue.index.map({
    0: "Unmarried",
    1: "Married"
})

marital_revenue.plot(kind='bar')

plt.xlabel("Marital Status")
plt.ylabel("Total Revenue")
plt.title("Revenue by Marital Status")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

```

#### *Insight: Unmarried customers contribute higher total revenue (₹6.21 Cr) and place more orders (16,249) compared to married customers; however, the average spend per order is nearly similar (₹3,823 vs ₹3,761), indicating that the revenue difference is primarily driven by order volume rather than significantly higher spending behavior*



### Insight Summary

 - The 26–35 age group dominates the customer base and leads in both total revenue (₹4.28 Cr+) and order volume, establishing it as the core revenue-driving segment.
 - Revenue is geographically concentrated, with UP leading at ₹1.9 Cr, and strong contributions from Maharashtra, Karnataka, and Delhi, while the Central zone emerges as the highest-performing region at ₹4.16 Cr.
 - IT professionals generate the highest revenue among occupations, followed by Healthcare, Aviation, and Banking, indicating stronger purchasing power within high-income segments.
 - Food is the dominant product category across both genders, significantly outperforming others, with lifestyle categories such as Clothing, Electronics, and Footwear forming the next major contributors.
 - Gender-based purchasing behavior is largely similar, with minor differences in category preference (females leaning toward Footwear, males toward Electronics).
 - Unmarried customers contribute higher total revenue and order volume; however, average spend per order remains nearly identical across marital status, suggesting revenue differences are volume-driven rather than spending-intensity driven.


### Strategic Recommendations

- Focus marketing on 26–35 age segment.
- Strengthen Central zone operations due to dominant revenue contribution.
- Expand Food category offerings as it is the primary revenue driver.
- Target unmarried segment with retention campaigns due to higher order volume.
